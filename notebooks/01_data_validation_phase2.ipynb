```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/uganda_fake_news_v2.csv",
    skipinitialspace=True
)

print(df.shape)
df.head()

```

```python


# Check dataset shape
df.shape
```

```python
# 
duplicates = df[df.duplicated(subset=["text"], keep=False)]
duplicates.sort_values("text").head(20)

```

```python


#Check for missing values
df['label'].value_counts()

df.duplicated().sum()
df = df.drop_duplicates()

df['text_length'] = df['text'].str.split().str.len()

df['text_length'].describe()
```

```python
# 
df.duplicated(subset=["text"]).sum()
```

```python
df = df.drop_duplicates(subset=["text"])
```

```python
df.duplicated(subset=["text"]).sum()
```

```python
dup_counts = df["text"].value_counts()
dup_counts[dup_counts > 1]
```

```python
# Confirming dataset shape
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# checking no structural null explosion
df.isnull().sum()
```

```python


# confirming platform types
df["platform_type"].value_counts()
```

```python
# checking label distribution/ text sanity check
df.sample(5)[["text", "label"]]
```
